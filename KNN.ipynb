```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoding=LabelEncoder()
```

```python
x.Gender=encoding.fit_transform(x.Gender)
```

```python
x
```

```python
x.info()
```

```python
skf=StratifiedKFold(n_splits=5)
```

```python
for train_index,test_index in skf.split(x,y):
  x_train,x_test=x.iloc[train_index],x.iloc[test_index]
  y_train,y_test=y.iloc[train_index],y.iloc[test_index]
```

```python
# In your code, you are working with a dataset to predict iPhone purchases based on gender, age, and salary. Let's say the 'Purchase Iphone' column has two classes: 0 (No purchase) and 1 (Purchase). If the dataset is imbalanced (e.g., more samples with 'No purchase'), using StratifiedKFold will ensure that each fold has a similar proportion of 'Purchase' and 'No purchase' samples. This helps to ensure that the model is trained and evaluated on a representative sample of both classes.

# Without StratifiedKFold, a simple random split might result in some folds having a disproportionate number of 'No purchase' samples, leading to a model that is biased towards predicting 'No purchase'. StratifiedKFold addresses this issue by maintaining the class proportions across the folds, improving the model's generalization ability.

```

feature scaling

```python
scale=StandardScaler()
```

```python
x_train=scale.fit_transform(x_train)
x_test=scale.transform(x_test)# converts into 0's and 1's for accurate results
```

choosing the model ,model selection

```python
# newX.head()
```

```python
log=LogisticRegression()
knn= KNeighborsClassifier(n_jobs=5)
```

Fitting the data

```python
log.fit(x_train,y_train)
knn.fit(x_train,y_train)
```

Testing the model

```python
y_knn_pred=knn.predict(x_test)
```

```python
y_log_pred=log.predict(x_test)
```

```python
new_df=pd.DataFrame({'Actual':y_test,'Predicted':y_knn_pred})
```

```python
accuracy_score(y_test,y_log_pred) # knn given more accuarcy i.e 86% than 70 0.7 %
```

```python
confusion_matrix(y_test,y_knn_pred)
```

```python
# accuracy from above confusion matrix is (42+27)/sum of all i.e 80 => 68/80=approx 86% is accuracy
lis=[i for i in range(3,100)if i%2!=0]
```

```python
acc=[]
dic={}
for i in lis:
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train,y_train)
  y_knn_pred=knn.predict(x_test)
  acc.append(accuracy_score(y_test,y_knn_pred))
  dic[i]=accuracy_score(y_test,y_knn_pred)
print( max(acc))
```
